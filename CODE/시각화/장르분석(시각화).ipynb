### 2010년~2022년 연도차트 장르별 분포

```python
# 한곡에 여러개 장르가 포함될 경우 포함되는 모든 장르에 곡 수 +1 증가시킴
split_lst = [val.split(', ') for val in df['Genre']]
sum_lst = sum(split_lst, [])  # 리스트 내 리스트 합치기
```

```python
from collections import defaultdict

genre_dict = defaultdict(int)

for one_genre in sum_lst:
    genre_dict[one_genre] += 1
```

```python
genre_dict
```

```python
len(genre_dict.keys())
```

```python
genre_dict.keys()
```

```python
genre_dict.values()
```

```python
genre_cnt = pd.DataFrame({'cnt' : genre_dict.values()},
             genre_dict.keys())
```

```python
genre_cnt.head(3)
```

```python
genre_cnt = genre_cnt.sort_values('cnt', ascending=False)
genre_cnt.head(3)
```

```python
genre_cnt
```

```python
plt.rcParams['font.family'] = 'NanumGothic'

# 시각화 설정
plt.figure(figsize=(10, 8))  # 그림 크기 설정 (선택사항)

# 막대 그래프로 장르 분포 시각화
# genre_counts.plot(kind='bar', color='skyblue')
plots = genre_cnt.plot(kind='bar', color='skyblue')
plt.title('2010년~2022년 연도차트 장르별 분포')
plt.xlabel('장르')
plt.ylabel('개수')


# 값 레이블 달기
for bar in plots.patches:

  # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # free space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment
    plots.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=7, xytext=(0, 4),
                   textcoords='offset points')


# 그래프 표시
plt.xticks(rotation=90)  # X축 레이블 회전 (선택사항)
plt.tight_layout()
plt.show()
```

### 연도별 장르 개수 누적 막대그래프

```python
tmp = [[year, val.split(', ')] for val, year in zip(df['Genre'], df['year'])]

tmp_lst = []
for e, genre_lst in tmp:
    for genre in genre_lst:
        tmp_lst.append([e, genre])
```

```python
tmp_lst[:5]
```

```python
year_lst = [year for year, genre in tmp_lst]
genre_lst = [genre for year, genre in tmp_lst]

year_genre_df = pd.DataFrame({'year':year_lst,
              'genre':genre_lst})
```

```python
year_genre_df.head(3)
```

```python
# 연도별로 그룹화하여 장르 카운트
yearly_genre_counts = year_genre_df.groupby(['year', 'genre']).size().unstack(fill_value=0)
df_freq = yearly_genre_counts.div(yearly_genre_counts.sum(axis=1), axis=0)*100

# 시각화 설정
plt.figure(figsize=(24, 16))  # 그림 크기 설정

# 연도별 장르 분포 시각화 , Set3 , tab20c
df_freq.plot(kind='bar', stacked=True, cmap='tab20c')
plt.title('인기차트 연도별 장르 비율 누적 막대그래프')
plt.xlabel('연도')
plt.ylabel('')


# 그래프 표시
plt.xticks(rotation=45)  # X축 레이블 회전 (선택사항)
plt.tight_layout()
plt.legend(title='장르', bbox_to_anchor=(1.05, 1), loc='upper left')  # 범례 위치 설정 (선택사항)

# color map 변경
plt.summer()

plt.show()
```
